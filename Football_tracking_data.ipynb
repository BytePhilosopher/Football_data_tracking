```python
import os

file_path = '/content/Football_data_tracking/main.py'

with open(file_path, 'r') as f:
    content = f.read()

# Replace the incorrect path with the correct one and fix the typo
modified_content = content.replace(
    'RAW_VIDEO = "data/raw/kaggle/168.mp4"',
    'RAW_VIDEO = "data/raw/168.mp4"'
)
modified_content = modified_content.replace(
    'PROCESSED_VIDEO = "data/processed/168_processeded.mp4"',
    'PROCESSED_VIDEO = "data/processed/168_processed.mp4"'
)

with open(file_path, 'w') as f:
    f.write(modified_content)

print("Modified main.py to use the correct video path and fixed processed video path typo.")
```

```python
import os

file_path = '/content/Football_data_tracking/src/preprocess.py'

with open(file_path, 'r') as f:
    content = f.read()

# Replace 'avc1' with 'mp4v'
# This regex handles cases where 'avc1' might be followed by a comment or other characters
modified_content = content.replace("cv2.VideoWriter_fourcc(*'avc1')", "cv2.VideoWriter_fourcc(*'mp4v')")

with open(file_path, 'w') as f:
    f.write(modified_content)

print("Successfully modified src/preprocess.py to use 'mp4v' codec.")
```

```python
import os

file_path = '/content/Football_data_tracking/src/tracker.py'

with open(file_path, 'r') as f:
    content = f.read()

# Replace the incorrect bbox access with .xyxy
# This assumes 'd' in the list comprehension is an ultralytics.engine.results.Boxes object
modified_content = content.replace('boxes = np.array([d["bbox"] for d in detections])', 'boxes = np.array([d.xyxy for d in detections])')

with open(file_path, 'w') as f:
    f.write(modified_content)

print("Modified src/tracker.py to correctly extract bounding box data.")
```
